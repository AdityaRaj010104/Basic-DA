### header parameter

```python
pd.read_csv('data.csv',header = 1) #covert first row into header columns
```

### use cols parameter

```python
pd.read_csv('data.csv',usecols = ['Socioeconomic Score','Grades']) # will select only mentioned columns
```

### squeeze parameters

```python
pd.read_csv('data.csv', usecols = ['Grades'],squeeze=True)
```

### skiprows/nrow parameter

```python
pd.read_csv('data.csv', skiprows = [0,2]) #will skip the row 0 and row 2
```

```python
pd.read_csv('data.csv', nrows = 400) #in case of large dataset we work on it by brekings into parts
```

### Encoding parameter

```python
pd.read_csv('data.csv', encoding = 'latin-1')
```

###skip bad lines

```python
pd.read_csv('data.csv',sep = ';', encoding = "latin-1", error_bad_lines = False )
```

### dtype parameter

```python
pd.read_csv('data.csv',dtype = {'Attendance (%)':int})
```

### handling dates

```python
pd.read_csv('data.csv',parse_dates = ['Date']) #if we have date column then it will covert it to actual date format
```

# EDA using bivarate analysis

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
titanic = pd.read_csv('train.csv')
```

```python
iris = sns.load_dataset('iris')
```

## 1 Scatterplot (numerical - numerical)

```python
sns.scatterplot(x = tips['total_bill'],y = tips['tip'],hue = tips['sex'],style = tips['smoker'],size = tips['size'])
```

## 2 Barplot (numrical - categorical)

```python
titanic.head()
```

```python
sns.barplot(x = titanic['Pclass'],y = titanic['Age'],hue = titanic['Sex'])
```

```python
sns.barplot(x = titanic['Pclass'],y = titanic['Fare'])
```

## 3 Boxplot (numerical - cagtegorical)

```python
sns.boxplot(x = titanic['Sex'],y = titanic['Age'],hue = titanic['Survived'],palette='coolwarm')
```

## 4 Distplot (numerical - categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'], hist = False)
sns.distplot(titanic[titanic['Survived']==1]['Age'], hist = False)
```

## 5 Heatmap (categorical - categorical)

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Pclass').mean(numeric_only = True)['Survived']*100 #kis class me kitne log bache
```

```python
colors = ['Red','Green','Blue']
(titanic.groupby('Pclass').mean(numeric_only = True)['Survived']*100).plot(kind = 'bar',color = colors)
```

```python
titanic.groupby('Sex').mean(numeric_only = True)['Survived']*100 #kis sex ke kitne log bache
```

```python
colors = ['Red','Green']
(titanic.groupby('Sex').mean(numeric_only = True)['Survived']*100).plot(kind = 'bar',color = colors)
```

## 6 Clustermap (categorical - categorical)

```python
pd.crosstab(titanic['SibSp'],titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived'])) #sibling spouse
```
